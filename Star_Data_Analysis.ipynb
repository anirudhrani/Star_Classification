# Univariate Analysis

### Temperature

```python
df['Temperature (K)'].describe()
```

```python
sns.distplot(df['Temperature (K)'])
```

```python
sns.boxplot(df['Temperature (K)'])
```

```python
t= (df['Temperature (K)'])
print('Lower Quartile (Q1) =',t.quantile(0.25),'\nUpper Quartile (Q3) =',t.quantile(0.75),
      '\nInter Quartile Range (IQR) =',t.quantile(0.75)-t.quantile(0.25)) 
t_q1=t.quantile(0.25)
t_q3=t.quantile(0.75)
t_iqr= t_q3-t_q1
print('Lower boundary value:', t_q1 - 1.5*t_iqr)
print('Upper boundary value:',t_q3 + 1.5*t_iqr)
```

```python
print('Higher outliers detected:- ',t[df['Temperature (K)']> 32622.375].count())
print('Lower outliers detected:- ',t[df['Temperature (K)']< -14222.625].count() )
```

```python
df[df['Temperature (K)']== t[df['Temperature (K)']> 32622.375].max()]
```
